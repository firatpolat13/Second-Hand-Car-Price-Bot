```python
import requests
from bs4 import BeautifulSoup

url = "https://www.arabam.com/ikinci-el/otomobil?page=1"
r = requests.get(url)

print(r.status_code)
r.content
```

```python
soup = BeautifulSoup(r.content, "lxml")
page = int(soup.find("div", attrs={"class":"listing-new-pagination cb tac mt16 pt16"}).find("span").text)
```

```python
ads = soup.find("tr", attrs={"class":"listing-list-item pr should-hover bg-white"}).text
ad_title = soup.find("td", attrs={"class":"listing-modelname pr"}).text
ad_car_year = soup.find("div", attrs={"class":" fade-out-content-wrapper"}).text
ad_listingprice = soup.find("a", attrs={"class":"listing-price"}).text
print(str(ad_title) + " Price: " + str(ad_listingprice) + " Year: " + str(ad_car_year) )
```

```python
count = 0
page_no = 1
for i in range(1,page):
    url = "https://www.arabam.com/ikinci-el/otomobil?page="+str(i)
    r = requests.get(url)
    soup = BeautifulSoup(r.content, "lxml")
    all_ads = soup.find_all("tr", attrs={"class":"listing-list-item pr should-hover bg-white"})
    
    print("Page No: " + str(page_no))
    print("*"*50)
    page_no += 1
    for ad in all_ads:
        print("Brand: " + ad.find("td", attrs={"class":"listing-modelname pr"}).text.split(" ")[0])
        print("Model: " + ad.find("td", attrs={"class":"listing-modelname pr"}).text.split(" ")[1])
        print("Description: " + str(ad.find("td", attrs={"class":"listing-modelname pr"}).text.split(" ")[2:]))
        print("Year: " + str(ad.find("div", attrs={"class":" fade-out-content-wrapper"}).text))
        print("Price: " + ad.find("a", attrs={"class":"listing-price"}).text)
        print("*"*30)
        count += 1
print("Total ads:" + str(count))
```
